## 异常检测分析  

-  **什么是异常检测**  
> 定义： 孤立点、离群、与众不同、超出范围、不在预期...  
> 原因： 测量误差、记录误差、随机、非随机其他因素导致...  
> 应用： 故障检测、网络入侵、欺诈分析、医疗药物...  

-  **异常vs正常**  
> 一个人的异常可能是另一个人的信号  --推断来自抽样，应用需要谨慎  
作为运动青年和普通青年的差异：

-  **异常检测方法**  
> 统计：非随机性   
> 距离：组织之外   
> 密度：且看身边  
> 其他：序列、偏差...
